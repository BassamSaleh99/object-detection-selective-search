```python
for a, i in enumerate(data.values):
    img = plt.imread(f'D:\\projects\\Computer Vision Project\\Computer Vision\\data\\training_images\\{i[0]}')
    print(img.shape)
    
    img_1 = img.copy()
    
    plt.figure()
    plt.imshow(img)
    
    xmin = int(i[1])
    ymin = int(i[2])
    xmax = int(i[3])
    ymax = int(i[4])
    
    cv2.rectangle(img_1, (xmin, ymin), (xmax, ymax), (255, 0, 0), 2)
    
    plt.figure()
    plt.imshow(img_1)
    
    if a == 2:
        break
```
